## Enunciat
Crea un aplicatiu en python que et mostri un menú que permeti escollir jugar a un dels següents diferents jocs:
- Adivina número: L’usuari tindrà tres intents per adivinar un número aleatori del 1 al 10. Per cada error es mostrarà per pantalla si el número és major o menor a l’indicat.
- Pedra - Paper - Tisores: Guanyador el primer que arriba a 3 punts.
- El ‘Penjat’ - ‘Ahorcado’: 
    - El programa partirà d’una llista de 30 paraules d’entre 3 i 7 lletres obtingudes d’un fitxer extern.
    - Aleatòriament es seleccionarà una de les 30 paraules.
    - L’usuari disposarà de número de lletres de la paraula * 2 intents per adivinar-la.
    - S’anirà mostrant abans de cada intent el tauler típic: R _ _ _ _

```python
# Importaciones
import random # importación de la librería random
```

```python

#Funciones
def esEntero(numero): # función que comprobará que el número que se le pase sea entero
    try:
        int(numero) # se convierte la variable pasada en la función como entero
    except ValueError: # Si la variable pasada no es un número saltará la excepción ValuError y la función retornará False
        print("El número no es entero inserte otro número porfavor")
        return False
    else: # si no hay error significa que el número es un entero por lo tanto la función retornará True
        return True

def escogerOpcion(): # función que te permitirá seleccionar una opción entre la a y la c
    opionValida = False
    while not opionValida:
        opcionInput = input("Seleccione la opción que deseas:") 
        if opcionInput in ["a", "b", "c"]: # comprobará que la opción sea a b c si no es una de estas 3 volverá a solicitar otra vez por pantalla
            opcionValida = True
            break
    return opcionInput

def ganaUser(eleccionUser, eleccionBot):
        if((eleccionUser == 'tijeras' and eleccionBot == 'papel') or (eleccionUser == 'papel' and eleccionBot == 'piedra')  or (eleccionUser == 'piedra' and eleccionBot == 'tijeras')):
            return True
        else:
            return False
def checkNumber(): # Función que pide un número y compruba que este comprendido entre el 1 y el 10
    isNumber = False
    while not isNumber:
        numeroInput = input("Cual crees que és el número")
        if esEntero(numeroInput): # Comprueba que el número sea entero 
            numeroInput = int(numeroInput) # si este número es entero se convertira al tipo int 
            if numeroInput in range(1, 10): # comprobará que el número esté en el rango del 1 al 10 y si no lo está volverá a solicitar otro
                isNumber = True 
                break
            else:
                print("Número fuera de rango tiene que ser un número entre el 1 y el 10")
    return numeroInput

def comprobarMenorMayor(numeroAdivinar, numeroUser): # Función que comprobará si el número insertado es menor o mayor al número a adivivinar
    if(numeroAdivinar < numeroUser):
            print("Tu número es mayor al número a adivinar")
            return False
    elif(numeroAdivinar > numeroUser):
            print("Tu número es menor al número a adivinar")
            return False
    else: # Si los dos números son iguales la funcion retornará True 
        return True
    
def setTablero(palabra):
    tablero = ['_'] * len(palabra) # se definirá un tablero con tantas _ como mida la palabra adivinar
    return tablero

def updateTablero(tablero,letra,palabra):
    if letra in palabra: # Comprobar que la letra está en la palabra adivinar
        for indice, letra_palabra in enumerate(palabra): # en caso de que la letra se encuentre en el tablero se pintará la parte correspondinte al tablero
            if letra == letra_palabra:
                tablero[indice] = letra
    
    return tablero
def setLetra(tablero):
    letraImbalida = True
    while letraImbalida:
        letraUser = input('Introduce una letra entre la a y la z ').lower()
        letraImbalida = not ('a' <= letraUser <= 'z' and len(letraUser) == 1)
        if letraImbalida :
            print(f"Maal el caracter '{letraUser}' es imbalida ha de comprenderse entre la a y la z y solo puede ser 1 carácter")
        else:
            letraImbalida = letraUser in tablero
            if letraImbalida:
                print("Letra repetrida")
    return letraUser
def checkResult(tablero):
    return '_' not in tablero
```

```python
print("Menú de selección de juego que se desee jugar elegir entre las opciones a, b y c")
print("- a: Adivina número")
print("- b: Piedra - Papel - Tijeras")
print("- c: El ‘Ahorcado’")
opcion = escogerOpcion()
```

```python
# Si se elige la opción a -> jugar a adivina el número
if (opcion == "a"):
    resultado = False # Definición del resultado por defecto será False ->
    print("Has escollit jugar a Adivina el numero: tindràs tres intents per adivinar un número aleatori del 1 al 10.")
    numeroAdivinar = random.randrange(1, 10) #elige un número aleatorio del 1 al 10
    numBucle = 0
    while numBucle < 3: # Iniciamos un bucle con 3 vueltas como máximo (numero de intentos)
        numeroUser = checkNumber() # Pedirá y comprobará que el número que pase el usuario sea correcto
        if(not comprobarMenorMayor(numeroAdivinar, numeroUser)): # compración si el número es mayor o menor al insertado por el usuario
            print("Por favor inserte otro número")
            numBucle+=1
        if(numeroUser == numeroAdivinar): # En caso de que los dos números sean iguales resultado será True que significa que ha ganado el usuario
            resultado = True
            
            break
    if(resultado):
        print("Has ganado en hora buena, una chapita para ti!")
    else:
        print(f"Muuuuuuuuuu mal has pedido el número era: {numeroAdivinar}")
```

```python
# Si el usuario escoge la opción b -> jugará pidra papel tijeres
if (opcion == "b"):
    print("Has seleccionado jugar a piedra papel o tijeras. Jugarás contra mi eligirás piedra papel o tijeras y yo sacaré una aleatorio el que gane suma 1 punto a su marcador el primer jugador en llegar a 3 puntos gana.")
    puntosUser = 0 # Definición de los puntos del usuario iniciales
    puntosBot = 0 # Definición de puntos de la máquina que jugará contra el usuario
    posibilidades = ["piedra", "papel", "tijeras"] # definición de las posibilidades que puede tener el usuario para jugar
    while puntosUser < 3 and puntosBot < 3: # si comprobará el primero que llegue a 3 puntos y hasta que uno de los 2 no llegue a 3  puntos no se decidirá el jugador
        eleccionBot = random.choice(posibilidades) # El bot eligirá una posibilidad aleatoria de las antes definidas
        eleccionUser = input("Que elijes? ") # Se pedirá al usuario que elige
        while eleccionUser not in posibilidades: # en caso de que la opcion que ha eligido el usuario no esté comprendida entre las posibilidades si le volverá a pedir una correcta
            print("Te has equivocado tienes que elegir una palabra entre piedra papel o tijeras")
            eleccionUser = input("Que elijes? ")
        print(f"Yo saco {eleccionBot}")
        if((eleccionUser == eleccionBot)): # Se comprobará el resultado de la eleccion del usuario comparando con la aleatoria antes definida
            print("Hemos empatado siguiente ronda")
        elif(ganaUser(eleccionUser, eleccionBot)):
            print("Ganas tú, sumas 1 punto")
            puntosUser+=1
        else:
            print("Gano yo, sumo 1 punto")
            puntosBot+=1
        
        print("Recuento de puntos:")
        print(f"Tu llevas: {puntosUser} puntos y yo llevo: {puntosBot} puntos")
    if(puntosUser == 3):
        print("Has llegado a 3 puntos ganas tú en hora buena un pn para ti!!")
    if(puntosBot == 3):
        print("Gano yo jdt!")
```

```python
# Si el usuario elige la opción c -> jugará al ahorcado
if (opcion == "c"):
    print("El ‘Ahorcado’")
    lista = []
    for linea in open('palabras.txt'): # se recorrerá un fichero que tendrá todas las palabras que pueden salir en el ahorcado y las almacenará en una lista
        lista.append(linea.strip())
    
    palabraAdivinar = random.choice(lista) # De lista generada por el archivo se eligira una palabra aleatoria para que el usuario adivine
    tablero = setTablero(palabraAdivinar.lower()) # Se definirá un tablero con las lineas correspondientes a las palabras por adivinar
    palabraFinal = ""
    resultado = False
    numeroIntentos = len(palabraAdivinar) * 2  # Se calcula el número de intentos máximos a realizar por el usuario
    print(f"Numero intentos: {numeroIntentos}")
    for intento in range(numeroIntentos): # Se repitará esta acción dependiendo del número de intentos que haya
        letraUser = setLetra(tablero) # Se le pedirá al usuario que reporte una letra entre la a y la z y si no es comprendida entre estas le pedirá de nuevo
        tablero= updateTablero(tablero, letraUser, palabraAdivinar) # a partir de la letra que ha pasado el usuario se comprobará que esté o no en el tablero y se añadirá
        print(f"Intento número {intento+1}, así queda el tablero después de insertar la letra {letraUser}:")
        print(" ".join(tablero))
        if checkResult(tablero) : # se comprobará si se ha ganado o no
            break
        intento+=1
    resultado = checkResult(tablero) # se guardará el resultado
    if resultado : 
        print("Bieeeeeeeeeeeen Has ganado")
    else:
        print("Pa junio crack. La palabra era:")
        print(palabraAdivinar)
```
